```python
def load_data(csv_path):
    df_merged = pd.read_csv(csv_path, index_col=[0])
    return df_merged
```

```python
df_merged = load_data(csv_path=CSV_PATH)
```

```python
ORDER_OF_COLUMNS = [
    'pk_cid',
    'pk_partition',
    'entry_date',
    'entry_channel',
    'entry_channel_id',
    'segment',
    'country_id',
    'region_code',
    'gender',
    'age',
    'age_group',
    'deceased',
    'salary',
    'active_customer',
    'em_acount',
    'em_account_p',
    'emc_account',
    'em_account_pp',
    'short_term_deposit',
    'loans',
    'mortgage',
    'funds',
    'securities',
    'long_term_deposit',
    'credit_card',
    'debit_card', 
    'payroll',
    'payroll_account', 
    'pension_plan'
]
```

```python
def preprocessing_data(df_merged):
    
    df_merged["entry_channel_id"] = OrdinalEncoder().fit_transform((df_merged["entry_channel"].values).reshape(-1,1))

    df_merged = df_merged[ORDER_OF_COLUMNS]

    return df_merged
```

```python
df_merged = preprocessing_data(df_merged=df_merged)
```
