Generamos una lista de producto :

```python
products = ['short_term_deposit',
 'loans',
 'mortgage',
 'funds',
 'securities',
 'long_term_deposit',
 'em_account_pp',
 'credit_card',
 'payroll',
 'pension_plan',
 'payroll_account',
 'emc_account',
 'debit_card',
 'em_account_p',
 'em_acount']
```

Observamos que son columnas que funcionnan de manera boolean :
0 => el cliente no tiene el producto
1 = el cliente tiene el producto

```python
#Observamos que los Dtype de las columnas no son optimizados

df.info()
```

Vamos a observar si tenemos nulos para poder convertir las columnas en int8

## Check Nulos

```python
df.isnull().sum()
```

```python
#A los nulos de pension_plan y payroll son 61 valores nulos en un dataset de más de 5 milliones de filas. Los pondremos la Moda.
```

```python
for col in ['pension_plan', 'payroll']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

## Convertir Columnas de Productos en int8

```python
df[products] = df[products].astype('int8')
df.info()

#Acabamos de didivir en el dataset por 4 de tamaño. De 800mo a 221mo.
```

## Check Duplicados

```python
df[df.duplicated()]

#NO TENEMOS DUPLICADOS
```

```python
#Cambiamos tambien pk_partition a tipo de fecha :
df['pk_partition'] = pd.to_datetime(df['pk_partition'])

df.info()
```

## EDA

### Observación por producto del porcentaje de contratación:

```python
df_gb_product = df.groupby('pk_partition')[products].sum()
df_gb_product
```

```python
for col in df_gb_product:
  sns.countplot(df[col])
  plt.title(col)
  plt.show()
```

Menos el productos "em_account", los productos tienen mucho más 0 que 1. Clientes que no tienen los productos contratados.

```python
#Observación en porcentaje :
for col in df[products]:
    print(df[col].value_counts(normalize=True))
```

### Observación por producto a traves del tiempo :

```python
for col in products:
    plt.figure()
    plt.plot(df_gb_product.index, df_gb_product[col])
    plt.xlabel('Fecha')
    plt.ylabel('Clientes')
    plt.title(f'Clientes de {col} por fecha')
    plt.xticks(rotation=75)
    plt.show()
```

Observación Global de todos los productos :

```python
for col in products:
    plt.plot(df_gb_product.index, df_gb_product[col], label=col)

plt.xlabel('date')
plt.ylabel('clientes')
plt.xticks(rotation=75)
plt.legend()
plt.show()
```

Vemos que el producto "em_account" no permite de ver la evolución de los otros produtos :

## Creación de una columna Tenencia (numeros de productos contratados por cliente):

Sumar por cada cliente el numero de producto que tiene contratados :

Definimos la nueva columna 'tenencia'

```python
df['tenencia'] = 0
```

Hacemos la sum de los productos en la columna 'tenencia'

```python
df['tenencia'] = df[products].sum(axis=1)
```

Controlamos los resultados

```python
df['tenencia'].value_counts().sort_values(ascending=False)
```
