## 1 A crear la tabla de ventas de los productos

```python
# Cambiamos la columna 'pk_partition' en datetime
df['pk_partition'] = pd.to_datetime(df['pk_partition'])

# Sacamos la lista de columna
product_columns = ['loans', 'mortgage', 'funds', 'securities', 'long_term_deposit', 'em_account_pp',
                   'credit_card', 'payroll', 'pension_plan', 'payroll_account', 'emc_account',
                   'debit_card', 'em_account_p', 'em_acount']

# Agrupamos por 'pk_cid' y calculamos la diferencia entre 2 meses consecutivos
df_diff = df.set_index('pk_partition').groupby('pk_cid')[product_columns].resample('M').ffill().diff()

df_diff.head()
```

Ahora vamos a generar la tabla de venta

De momento tenemos : 
1. 0 si 2 rows are identicas (no hubo venta o baja)
2. 1 si hay una venta
3. -1 si hay una baja

```python
# Por cada row y cada producto creamos una columna de venta "sale_{column_name}":
for col in product_columns:
    df_diff[f"sale_{col}"] = df_diff[col].apply(lambda x: 1 if x == 1 else 0)

print(df_diff)
```

Los pk_partition han sido transformando en el resample, vamos a transformarlos para que los días sean todos igual a 28

```python
df_diff.reset_index(inplace=True)
```

```python
df_diff['pk_partition'] = df_diff['pk_partition'].astype(str).str.replace(r'-30$', '-28').str.replace(r'-31$', '-28')
```

Vamos a proceder al merge

```python
#primero sacamos la lista de columna que nos interesa

col_merge = ['pk_cid', 
 'pk_partition', 
 'sale_loans',
 'sale_mortgage',
 'sale_funds',
 'sale_securities',
 'sale_long_term_deposit',
 'sale_em_account_pp',
 'sale_credit_card',
 'sale_payroll',
 'sale_pension_plan',
 'sale_payroll_account',
 'sale_emc_account',
 'sale_debit_card',
 'sale_em_account_p',
 'sale_em_acount']
```

```python
df_diff['pk_partition'] = pd.to_datetime(df_diff['pk_partition'])
```

```python
df_merged = df.merge(df_diff[col_merge], on=['pk_cid', 'pk_partition'])
```
